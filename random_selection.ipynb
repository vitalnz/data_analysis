## Random selection process using python

## Importing libraries and do calculations

```python
# Importing libraries
import pandas as pd
from IPython.display import display, HTML
# Reading data
path = "/content/drive/MyDrive/data_analysis/data/tracers2.csv"
df = pd.read_csv(path)
# Calculate to total sample size
tot = df.shape[0]
# Set sample size, or percentage to be used. User can change to see the results
perc = 0.35
sample_size = tot*perc
# Random selection using sample_size
selected = df.sample(n=int(sample_size))
selected_html = selected.to_html(index=False)
```

## Printing results

```python
print()
print('The size of the list before sampling is {}'.format(tot))
print()
print('The selected sample of randomly selected {} the following table:'.format(int(sample_size)))
print()
display(HTML(selected_html))
```
